```python
import numpy as np
import matplotlib.pyplot as plt

def linear_approximation(x, y):

    A = np.vstack([x, np.ones(len(x))]).T
    b, a = np.linalg.lstsq(A, y, rcond=None)[0]
    return a, b


np.random.seed(42)
x = np.arange(-20, 20)
y = 0.5 * x**3 + 1.5 * x**2 - 2.5 * x + 69 + np.random.normal(scale=200, size=len(x))

a, b = linear_approximation(x, y)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Dane rzeczywiste', alpha=0.6)
plt.plot(x, a + b * x, color='red', label=f'Funkcja aproksymujca: f(x) = {a:.2f} + {b:.2f}x')
plt.title('Aproksymacja liniowa redniokwadratowa')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()


a, b




def calculate_errors(x, y, a, b):
  y_approx = a + b * x
  absolute_error = np.abs(y-y_approx)
  relative_error = np.abs((y - y_approx) /y) * 100
  return absolute_error, relative_error

absolute_error, relative_error = calculate_errors(x, y, a, b)

mean_absolute_error = np.mean(absolute_error)
mean_relative_error = np.mean(relative_error)

mean_absolute_error, mean_relative_error
```
