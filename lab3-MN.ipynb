```python
import numpy as np

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def tsin(x, terms=10):
    sin_approx = 0
    for n in range(terms):
        sin_approx += ((-1)**n * x**(2 * n + 1)) / factorial(2 * n + 1)
    return sin_approx

def tcos(x, terms=10):
    cos_approx = 0
    for n in range(terms):
        cos_approx += ((-1)**n * x**(2 * n)) / factorial(2 * n)
    return cos_approx

def texp(x, terms=10):
    exp_approx = 0
    for n in range(terms):
        exp_approx += x**n / factorial(n)
    return exp_approx

def ttan(x, terms=10):
    return tsin(x, terms) / tcos(x, terms)

x_test = np.pi / 4
print(tsin(x_test), tcos(x_test), texp(x_test), ttan(x_test))
```

```python
#zad4
import numpy as np

def blad_wzgledny(aprox, dokladne):
    return np.abs((aprox - dokladne) / dokladne) if dokladne != 0 else np.nan

def blad_bezwzgledny(aprox, dokladne):
    return np.abs(aprox - dokladne)

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def tsin(x, terms=10):
    sin_approx = 0
    for n in range(terms):
        sin_approx += ((-1)**n * x**(2 * n + 1)) / factorial(2 * n + 1)
    return sin_approx

def tcos(x, terms=10):
    cos_approx = 0
    for n in range(terms):
        cos_approx += ((-1)**n * x**(2 * n)) / factorial(2 * n)
    return cos_approx

def texp(x, terms=10):
    exp_approx = 0
    for n in range(terms):
        exp_approx += x**n / factorial(n)
    return exp_approx

def ttan(x, terms=10):
    return tsin(x, terms) / tcos(x, terms) if tcos(x, terms) != 0 else np.nan

def testuj_wspolczynniki_zmiennosci(f_approx, f_exact, zakres, opis):
    x_vals = np.linspace(zakres[0], zakres[1], 100)
    bledy_wzgledne = []
    bledy_bezwzgledne = []

    for x in x_vals:
        aprox = f_approx(x)
        dokladne = f_exact(x)
        bw = blad_wzgledny(aprox, dokladne)
        bb = blad_bezwzgledny(aprox, dokladne)
        if not np.isnan(bw):
            bledy_wzgledne.append(bw)
        bledy_bezwzgledne.append(bb)

    sr_wzgledny = np.mean(bledy_wzgledne) if bledy_wzgledne else None
    sr_bezwzgledny = np.mean(bledy_bezwzgledne) if bledy_bezwzgledne else None
    odch_std_wzgledny = np.std(bledy_wzgledne) if bledy_wzgledne else None
    odch_std_bezwzgledny = np.std(bledy_bezwzgledne) if bledy_bezwzgledne else None

    cv_wzgledny = (odch_std_wzgledny / sr_wzgledny) if sr_wzgledny else None
    cv_bezwzgledny = (odch_std_bezwzgledny / sr_bezwzgledny) if sr_bezwzgledny else None

    print(f"\n{opis}")
    print(f"Współczynnik zmienności błędu względnego: {cv_wzgledny:.2e}" if cv_wzgledny is not None else "Brak danych")
    print(f"Współczynnik zmienności błędu bezwzględnego: {cv_bezwzgledny:.2e}" if cv_bezwzgledny is not None else "Brak danych")

zakresy = [
    (tsin, np.sin, [-3 * np.pi, 0], "tsin(x) vs numpy.sin(x) [-3π, 0]"),
    (tsin, np.sin, [0, np.pi / 4], "tsin(x) vs numpy.sin(x) [0, π/4]"),
    (tsin, np.sin, [-3 * np.pi, 3 * np.pi], "tsin(x) vs numpy.sin(x) [-3π, 3π]"),
    (tcos, np.cos, [-3 * np.pi, 0], "tcos(x) vs numpy.cos(x) [-3π, 0]"),
    (tcos, np.cos, [0, np.pi / 4], "tcos(x) vs numpy.cos(x) [0, π/4]"),
    (tcos, np.cos, [-3 * np.pi, 3 * np.pi], "tcos(x) vs numpy.cos(x) [-3π, 3π]"),
    (texp, np.exp, [0, 20], "texp(x) vs numpy.exp(x) [0, 20]"),
    (texp, np.exp, [0, 1], "texp(x) vs numpy.exp(x) [0, 1]"),
    (ttan, np.tan, [-np.pi / 2, 0], "ttan(x) vs numpy.tan(x) [-π/2, 0]"),
    (ttan, np.tan, [0, np.pi / 2], "ttan(x) vs numpy.tan(x) [0, π/2]"),
    (ttan, np.tan, [-np.pi / 2, np.pi / 2], "ttan(x) vs numpy.tan(x) [-π/2, π/2]")
]

for f_approx, f_exact, zakres, opis in zakresy:
    testuj_wspolczynniki_zmiennosci(f_approx, f_exact, zakres, opis)
```
