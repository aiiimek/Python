```python
#zad 1
import numpy as np
import pylab as plt

xk = 0.1
xyv = []

xyv.append([np.cos(1/i)for i in np.arange(0,3.3,0.1) if i !=0])
xyv.append([np.exp(np.sin(i))for i in np.arange(0,3.3,0.1) if i !=0])
print(xyv)

print(len(xyv[0]))

plt.grid()
plt.plot(xyv[0],xyv[1],'k',xyv[0],'-r')
plt.show()

def interplin2p (x,xi,yi,xil,yil):
  return yi+(yil-yi)*(x-xi)/(xil-xi)

def interplinvect (x,xyv):
  xinterp = []
  yinterp = []
  for xk in x:
    N = len(xyv[0])
    for i in range (0,N-1):
      if (xk >= xyv[0][i]) and (xk <= xyv[0][i+1]):
        xinterp.append(xk)
        yinterp.append(interplin2p(xk,xyv[0][i],xyv[1][i],xyv[0][i+1],xyv[1][i]))
        i = i + 1
  return [xinterp,yinterp]

xin = [i for i in np.arange (0,3.3,0.08)]
xyinterp = interplinvect(xin,xyv)
print (xyinterp)

plt.grid()
plt.plot(xyinterp[0],xyinterp[1],'k',xyv[0],'-r')
plt.tight_layout()
plt.show()

print(len(xyinterp[0]))
print(len(xyinterp[1]))
print(len(xin))

```

```python
#zad2
import numpy as np
import pylab as plt

def interpolacja(x,y,xval):
  yval = 0
  products = 0
  for i in range(len(x)):
    products = y[i]
    for j in range(len(x)):
      if i != j:
        products *= (xval - x[j])/(x[i] - x[j])
    yval += products
  return yval

x = [i for i in np.arange(-np.pi, np.pi, 0.1)]
y = [np.sin(1/i) for i in np.arange (-np.pi, np.pi, 0.1)]

print('x=',x)
print('y=',y)
print()

print('interpolowane: xval = %.lf oraz yval = %.lf' % (1, interpolacja(x,y,1)))
print('wartosci dokladne: xval = %.lf oraz yval = %.lf' % (1, np.sin(1/1)))

xval = [i for i in np.arange(-np.pi, np.pi, 0.1)]
yval = []
for xv in xval:
  yval.append(interpolacja(x,y,xv))

plt.grid()
plt.plot(xval,yval,'-y')
plt.show()

x = [i for i in np.arange(-np.pi, np.pi, 0.1)]
y = [np.exp(np.cos(i)) for i in np.arange (-np.pi, np.pi, 0.1)]

print('x=',x)
print('y=',y)
print()
print('interpolowane: xval = %.lf oraz yval = %.lf' % (1, interpolacja(x,y,1)))
print('wartosci dokladne: xval = %.lf oraz yval = %.lf' %(1, np.exp(np.cos(1))))

xval = [i for i in np.arange(-np.pi, np.pi, 0.1)]
yval = []
for xv in xval:
  yval.append(interpolacja(x,y,xv))


plt.grid()
plt.plot(xval,yval,'-y')
plt.show()
```
