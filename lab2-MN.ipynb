```python
#zad3
import numpy as np
import warnings
warnings.simplefilter('ignore')

def suma(f,xp, xk, k):
    sum=0
    for i in np.arange(xp, xk, k):
      sum=sum+f(i)
    return sum

def iloczyn(f, xp, xk, k):
    il=1
    for i in np.arange(xp, xk, k):
      il=il*f(i)
    return il

def odwr(x):
    try:
      wynik=1.0/x
    except ZeroDivisionError:
      print("dzielenie przez zero :( ")
    return wynik

print("Dla funkcji sinus:")
print(suma(np.sin, -3.14, 3.14, 0.01))
print(iloczyn(np.sin, -3.14, 3.14, 0.01))
print("Dla funkcji 1/x: ")
print(suma(odwr, -3.14, 3.14, 0.01))
print(iloczyn(odwr, -3.14, 3.14, 0.01))
```

```python
#zad4
import numpy as np
import math
import matplotlib.pyplot as plt

def rysuj(f, xp, xk, krok):
    x = np.arange(xp, xk, krok)
    y = f(x)

    plt.plot(x, y)
    plt.ylabel('f(x)')
    plt.xlabel('x')
    plt.title('Wykres funkcji ')
    plt.show()
rysuj(np.sin, -3.14, 3.14, 0.01)
```

```python
#zad 5
import numpy as np
import math
import matplotlib.pyplot as plt

def rysuj(f, xp, xk, krok, opcja=1):
    x = np.arange(xp, xk, krok)
    y = f(x)
    plt.plot(x, y)
    plt.ylabel('f(x)')
    plt.xlabel('x')
    if opcja == 1:
        plt.plot(x, y, 'o')
        plt.grid(True)
    plt.show()
rysuj(np.sin, -3.14, 3.14, 0.01, opcja=1)

```
